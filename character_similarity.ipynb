# Import 

```python
import pandas as pd
import os
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# Load

```python
data_dir = "data"
df = pd.read_csv(os.path.join(data_dir, "character_attributes.csv"))
df.head()
```

```python
names, attributes = df.iloc[:, 0], df.iloc[:, 1:]
values = np.array(attributes)
```

```python
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

# Cosine Similarity

```python
distances = []
name = "チノ"
idx = names[names == name].index[0]
value = values[idx, :]
for name_ in names:
    idx_ = names[names == name_].index[0]
    value_ = values[idx_, :]
    if idx == idx_:
        continue
    else:
        distance = cos_sim(value.reshape(-1), value_.reshape(-1))
        distances.append((name_, distance))
distances.sort(key=lambda x: -x[1])
print(distances[:10])
```

# Correlation Coefficient

```python
distances = []
name = "チノ"
idx = names[names == name].index[0]
value = values[idx, :]
for name_ in names:
    idx_ = names[names == name_].index[0]
    value_ = values[idx_, :]
    if idx == idx_:
        continue
    else:
        distance = np.corrcoef(value.reshape(-1), value_.reshape(-1))[0, 1]
        distances.append((name_, distance))
distances.sort(key=lambda x: -x[1])
print(distances[:10])
```

# Euclidean Distance

```python
distances = []
name = "チノ"
idx = names[names == name].index[0]
value = values[idx, :]
for name_ in names:
    idx_ = names[names == name_].index[0]
    value_ = values[idx_, :]
    if idx == idx_:
        continue
    else:
        distance = np.linalg.norm(value - value_)
        distances.append((name_, distance))
distances.sort(key=lambda x: x[1])
print(distances[:10])
```
